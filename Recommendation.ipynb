# Feature Selection

```python
movies.columns
```

```python
movies=movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genre']
```

```python
movies
```

```python
new_data = movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

Now,we're gonna convert the tags into vector and using cosine to detect the similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english') # We're gonna remove a,an,the,...in English by using stop word
```

```python
cv
```

If your data set contains 10000 rows (each row represents a text), and the number of different words in the entire dataset does not exceed 10000, the dictionary will have a maximum of 10000 words.

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['title']=="Joker"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommendation(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommendation("Domino")
```
